```python
Q1 = df['평점'].quantile(0.25)
Q3 = df['평점'].quantile(0.75)
IQR = Q3 - Q1
```

```python
print(f"1사분위(Q1): {Q1}, 3사분위(Q3): {Q3}, IQR: {IQR}")
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
print(f"하한값: {lower_bound}, 상한값: {upper_bound}")
```

```python
outliers = df[(df['평점'] < lower_bound) | (df['평점'] > upper_bound)]
```
