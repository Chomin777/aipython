```python
fruits = ['사과', '바나나', '딸기'] #과일과 판매 수량
counts = [10, 15, 7]
colors = ['green', 'yellow', 'red']
plt.bar(fruits, counts, color=colors)
#색상 지정(사과는 초록, 바나나는 노랑, 딸기는 빨강)
#바 차트 생성
plt.title('과일 판매량', fontsize=16) #제목, 폰트 크기
plt.xlabel('과일 종류', fontsize=14) # x축 레이블, 폰트 크기
plt.ylabel('판매 수량', fontsize=14) # y축 레이블, 폰트 크기
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

plt.figure(figsize = (10, 7))
#scatterplot에서는 이상치를 쉽게 관측할 수 있다.
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris, s=100, alpha=0.8)
plt.title('Iris 데이터 산점도 (sepal length vs sepal width)', fontsize=16)
plt.xlabel('꽃받침 길이 (sepal length)', fontsize=14)
plt.ylabel('꽃받침 너비 (sepal width)', fontsize=14)
plt.legend(title='종류 (species)')
plt.grid(True)
plt.show()
```
