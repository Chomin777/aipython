```python
def bmi_function(H, W):
  bmi =  W / (H / 100)**2
  return bmi
```

```python
height = float(input("당신의 키를 입력하시오"))
weight = float(input("당신의 몸무게를 입력하시오"))
BMI = round(bmi_function(height, weight), 2)
print(BMI)
```
