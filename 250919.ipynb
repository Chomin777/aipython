## f-string (Formatted String Literal)

f-string은 Python 3.6부터 도입된 문자열 포맷팅 방법으로, 문자열 안에 변수나 표현식의 값을 쉽게 삽입할 수 있게 해줍니다. 문자열 앞에 `f` 또는 `F`를 붙이고, 문자열 안에 삽입하고 싶은 변수나 표현식을 중괄호 `{}` 안에 넣어 사용합니다.

**장점:**
* **가독성:** 문자열과 변수/표현식이 함께 있어 코드를 읽기 쉽습니다.
* **성능:** 다른 문자열 포맷팅 방법(예: `%` 연산자, `.format()` 메서드)보다 일반적으로 더 빠릅니다.
* **간결성:** 코드가 더 짧고 간결해집니다.

**기본 사용법:**

```python
name = "Alice"
age = 30
print(f"안녕하세요, 제 이름은 {name}이고 나이는 {age}세입니다.")
```

**표현식 사용:**
f-string 안에서는 간단한 Python 표현식도 사용할 수 있습니다.

```python
x = 10
y = 20
print(f"x + y = {x + y}")
```

**포맷 지정:**
콜론 `:`을 사용하여 포맷을 지정할 수 있습니다. 예를 들어, 소수점 자리를 제한하거나 숫자에 콤마를 추가할 수 있습니다.

```python
pi = 3.1415926535
print(f"원주율은 {pi:.2f} 입니다.") # 소수점 둘째 자리까지 표시

number = 1000000
print(f"큰 숫자는 {number:,} 입니다.") # 천 단위 콤마 추가
```

**날짜 및 시간 포맷팅:**
datetime 객체와 함께 사용하여 날짜와 시간을 포맷팅할 수 있습니다.

```python
import datetime
now = datetime.datetime.now()
print(f"현재 날짜 및 시간: {now:%Y-%m-%d %H:%M:%S}")
```

f-string은 Python에서 문자열을 다룰 때 매우 유용하고 자주 사용되는 기능입니다.

```python
#마지막 문제: 원의 반지름을 입력받아 원의 넓이를 구하는 프로그램
from math import pi
```

```python
radius = float(input('원의 반지름을 입력'))
print(f"입력한 반지름은 {radius}이고, 원의 넓이는 {pi * radius**2:.2f}입니다.")
```
