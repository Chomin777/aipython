```python
#선택 구조 문제: 점수범위에 따라 ABCD 학점 부여하기
score = int(input("점수를 입력 "))
if score > 100 or score < 0:
  print("점수가 잘못되었습니다. 다시 입력해주십시오")
elif score >= 90 :
  print("A")
elif score >= 70 :
  print("B")
elif score >= 60 :
  print("C")
else:
  print("D")
```

```python
#중첩 해답
score2 = int(input("점수를 입력 "))
if score2 >= 0 and score2 <= 100 :
  if score2 >= 90 :
    print("A")
  elif score2 >= 70 :
    print("B")
  elif score2 >= 60 :
    print("C")
  else:
    print("D")
else:
  print("잘못된 점수입니다")
```

```python
#반복구조 for 예제-주의: print도 들여쓰면 print 글 또한 5번 반복된다!
sum = 0
for a in range(1, 101) :
  sum = sum + a

print("1부터 5까지의 합은", sum, "입니다.")
```

```python
#반복구조 for 문제: 홀수합 짝수합 구하기 - 해답1
odd_sum = 0
even_sum = 0
for a in range(1, 101) :
  if a % 2 == 0:
    even_sum = even_sum + a
  else:
    odd_sum = odd_sum + a

print("홀수합:", odd_sum, " 짝수합:", even_sum)
```

```python
#반복구조 for 문제: 홀수합 짝수합 구하기 - 해답2
sum = 0
for a in range(1, 101, 2):
  sum = sum + a

print("1부터 100까지의 홀수 합은 ", sum, "입니다.")

sum = 0 #같은 변수를 쓸 경우, 리셋해줘야 한다!
for a in range(0, 101, 2):
  sum = sum + a

print("1부터 100까지의 짝수 합은 ", sum, "입니다.")
```

```python
#반복구조 while 예제
sum = 0
money = 100
day = 1

while sum < 50000:
  sum = sum + money
  money = money + 100
  day = day + 1

print("총", day, "일이 걸립니다.")
```

```python
#반복구조 while 문제: 숫자를 입력받아서 총합을 구하는 프로그램을 작성하시오. 단 0을 입력하면 종료되고 최종 총합을 출력한다.
sum = 0
num = int(input("숫자입력: ")) # 첫 번째 숫자 입력

while num != 0 :
  sum = sum + num
  num = int(input("숫자입력: ")) # 다음 숫자 입력

print("최종 총합: ", sum)
```
